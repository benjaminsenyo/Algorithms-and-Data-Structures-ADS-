```python
def sum_of_two(nums, target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if i +j ==target:
                re = print(f" {i} and {j} and {nums[i]} and {nums[j]}")
                return re
```

```python
hey = [1,2,3,4,5]
sum_of_two(hey, 3)
```
